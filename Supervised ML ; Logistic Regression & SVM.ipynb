# DATA conversion in numericals

```python
from sklearn.model_selection import train_test_split
import sklearn.linear_model as lm
import sklearn.preprocessing as sp
```

```python
#separate features and labels
X=np.c_[data.iloc[: ,0:7]]
X
Y=data.iloc[:,7]
Y
```

# 1.vectorization


```python
encoder=sp.OneHotEncoder()
x_1= encoder.fit_transform(np.c_[data["Color"]]).toarray()
x_1


```

```python
x_2= encoder.fit_transform(np.c_[data["Eye_color"]]).toarray()
x_2
```

```python
x_3= encoder.fit_transform(np.c_[data["Moustache"]]).toarray()
x_3
```

```python
x_4= encoder.fit_transform(np.c_[data["Tail"]]).toarray()
x_4
```

# 2. label encoding

```python
encoder= sp.LabelEncoder()
Y_1= encoder.fit_transform(Y).reshape(-1,1)
Y_1
```

```python
a=pd.DataFrame(x_1)

a
```

```python
b=pd.DataFrame(x_2)
b
```

```python
c=pd.DataFrame(x_3)
c
```

```python
d=pd.DataFrame(x_4)
d
```

# Normalization

```python
height=data["Height"]
height
hgt_max= height.max()
hgt_max
for i in height:
    hgt_rng= height/hgt_max
hgt_rng
              
```

```python
weight=data["Weight"]
weight
wgt_max= weight.max()
wgt_max
for i in weight:
    wgt_rng= weight/wgt_max
wgt_rng
              
```

```python
legs=data["Legs"]
legs
leg_max= legs.max()
leg_max
for i in legs:
    leg_rng= legs/leg_max
leg_rng
            
```

# Dataframe ready for training

```python
X_1=pd.concat([a,b,hgt_rng,leg_rng,c,d,wgt_rng],axis=1)
X_1
```

```python
train_x,test_x,train_y,test_y= train_test_split(X_1, Y_1)
```

# ML Algorithm; logistic Regression

```python
model=lm.LogisticRegression()
model.fit(train_x,train_y)
pred_y=model.predict(test_x)
import sklearn.model_selection as ms
from sklearn.metrics import confusion_matrix
y_train_pred = ms.cross_val_predict(model, train_x, train_y)
pred_y=model.predict(test_x)
conf_mx = confusion_matrix(test_y, pred_y)
conf_mx = confusion_matrix(train_y, y_train_pred)
conf_mx
```

# ML Algorithm; SVM (kernel= linear)

```python
import sklearn.svm as sv
import sklearn.metrics as sm
model=sv.SVC(kernel='linear')
model.fit(train_x,train_y)
pred_y=model.predict(test_x)
print(sm.classification_report(test_y,pred_y))
import sklearn.model_selection as ms
from sklearn.metrics import confusion_matrix
y_train_pred = ms.cross_val_predict(model, train_x, train_y)
conf_mx = confusion_matrix(train_y, y_train_pred)
conf_mx
```

# SVM (kernel= degree 5)

```python
import sklearn.svm as sv
import sklearn.metrics as sm
model=sv.SVC(kernel='poly', degree= 5)
model.fit(train_x,train_y)
pred_y=model.predict(test_x)
print(sm.classification_report(test_y,pred_y))
import sklearn.model_selection as ms
from sklearn.metrics import confusion_matrix
y_train_pred = ms.cross_val_predict(model, train_x, train_y)
conf_mx = confusion_matrix(train_y, y_train_pred)
conf_mx
```
